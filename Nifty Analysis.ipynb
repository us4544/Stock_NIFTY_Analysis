```python
start = "2001-02-01"
end = '2021-09-09'
TATASTEEL = yf.download('TCS',start,end)
TCS = yf.download('INFY',start,end)
WIPRO = yf.download('WIPRO.NS',start,end)
```

```python
TATASTEEL['Open'].plot(label ='TataSteel', figsize = (15,7))
TCS['Open'].plot(label = "TCS")
WIPRO['Open'].plot(label = 'Wipro')
plt.title('Stock Prices of TATASTEEL,TCS,WIPRO')
plt.legend()
```

```python
TATASTEEL['Volume'].plot(label = 'Tatatsteel', figsize = (15,7))
TCS['Volume'].plot(label = "Tcs")
WIPRO['Volume'].plot(label = 'Wipro')
plt.title('Volume of Stock traded')
plt.legend()
```

```python
TATASTEEL['returns'] = (TATASTEEL['Close']/TATASTEEL['Close'].shift(1)) -1
TCS['returns'] = (TCS['Close']/TCS['Close'].shift(1))-1
WIPRO['returns'] = (WIPRO['Close']/WIPRO['Close'].shift(1)) - 1
TATASTEEL['returns'].hist(bins = 100, label = 'Tatasteel', alpha = 0.5, figsize = (15,7))
TCS['returns'].hist(bins = 100, label = 'Tcs', alpha = 0.5)
WIPRO['returns'].hist(bins = 100, label = 'Wipro', alpha = 0.5)
plt.legend()
```
